# Метрика

Перед решением любой задачи важно понимать, как будет оцениваться ваше решение. В данном случае мы используем стандартную для задачи классификации метрику ROC-AUC. Ее можно вычислить, используя только предсказанные вероятности и истинные классы без конкретного порога классификации + она раотает даже если классы в данных сильно несбалансированны (примеров одного класса в десятки раз больше примеров длугого). Именно поэтому она очень удобна для соревнований.

Посчитать ее легко:


```python
from sklearn.metrics import roc_auc_score

y_true = [
    0,
    1,
    1,
    0,
    1
]

y_predictions = [
    0.1,
    0.9,
    0.4,
    0.6,
    0.61
]

roc_auc_score(y_true, y_predictions)
```

```python
from sklearn.linear_model import LogisticRegression, LogisticRegressionCV
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, RobustScaler, LabelEncoder, OneHotEncoder
from sklearn.pipeline import make_pipeline
```

```python
# 1) Обработайте данные для того, чтобы к ним можно было применить LogisticRegression. Т.е. отнормируйте числовые признаки, а категориальные закодируйте с помощью one-hot-encoding'а.
# Разделение данных на тренировочную и валидационную выборки
X_train = data.drop(['Churn'], axis=1)
y_train = data['Churn']
features_train, features_valid, target_train, target_valid = train_test_split(
    X_train,
    y_train,
    train_size=0.8,
    random_state=12345)
features_train.shape, features_valid.shape, target_train.shape, target_valid.shape
# Replace spaces with NaN in numerical columns and then fill with 0
for col in num_cols:
    features_train[col] = features_train[col].replace(' ', float('nan')).fillna(0).astype(float)
    features_valid[col] = features_valid[col].replace(' ', float('nan')).fillna(0).astype(float)
    # Convert the column to numeric type to ensure compatibility with StandardScaler
```

```python
# Создаем ColumnTransformer для масштабирования числовых и кодирования категориальных признаков
from sklearn.linear_model import LogisticRegression, LogisticRegressionCV
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, RobustScaler, LabelEncoder, OneHotEncoder
from sklearn.pipeline import make_pipeline
from sklearn.compose import ColumnTransformer # Make sure to import ColumnTransformer
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), num_cols),
        ('cat', OneHotEncoder(drop='first', handle_unknown='ignore'), cat_cols)],
    remainder='passthrough')

# Создаем Pipeline, объединяющий предобработку и логистическую регрессию
pipeline = make_pipeline(
    preprocessor,
    LogisticRegression(random_state=12345, max_iter=1000)
)

# Определяем гиперпараметры
parametrs = {'logisticregression__C': [100, 10, 1, 0.1, 0.01, 0.001]}

# Создаем объект GridSearchCV для поиска наилучших параметров
grid_search_LR = GridSearchCV(pipeline, parametrs, cv=5, scoring='roc_auc')

# Обучаем модель
grid_search_LR.fit(features_train, target_train)

# Выведим наилучшие параметры и оценку AUC-ROC
print(grid_search_LR.best_params_)
print(grid_search_LR.best_score_)
```

```python
# Получаем вероятности и вычисляем AUC-ROC для валидационной выборки
probabilities_valid = grid_search_LR.predict_proba(features_valid)[:, 1]
LR_auc_roc = roc_auc_score(target_valid, probabilities_valid)

print("Логистическая регрессия:")
print("Значение AUC-ROC:", LR_auc_roc)
```

## Применение градиентного бустинга (2 балла)

Если вы хотите получить баллы за точный ответ, то стоит попробовать градиентный бустинг. Часто градиентный бустинг с параметрами по умолчанию даст вам 80% результата за 0% усилий.

Мы будем использовать catboost, поэтому нам не надо кодировать категориальные признаки. Catboost сделает это сам (в .fit() надо передать cat_features=cat_cols). А численные признаки нормировать для моделей, основанных на деревьях не нужно.

1) Разделите выборку на train/valid. Протестируйте catboost cо стандартными параметрами.

2) Протестируйте разные занчения параметров количества деревьев и learning_rate'а и выберите лучшую по метрике ROC-AUC комбинацию.

(Дополнительно) Есть некоторые сложности с тем, чтобы использовать CatBoostClassifier вместе с GridSearchCV, поэтому мы не просим использовать кроссвалидацию. Но можете попробовать)

```python
from catboost import CatBoostClassifier
cbst_simple = CatBoostClassifier(random_state=12345)

cbst_simple.fit(features_train, target_train, cat_features=cat_cols, verbose=False)

boost_preds_train = cbst_simple.predict_proba(features_valid)[:,1]
CBST_simple_roc_auc = roc_auc_score(target_valid, boost_preds_train)

print("CatBoost со стандартными настройками:")
print("Значение AUC-ROC:", CBST_simple_roc_auc)
```

```python
cbst = CatBoostClassifier(random_state=12345)

parametrs = {'learning_rate': [.3, .2, .1],
        'iterations': [1, 20, 100, 200]}

grid_search_catboost = GridSearchCV(cbst, parametrs, cv=5, scoring='roc_auc')

grid_search_catboost.fit(features_train, target_train, cat_features=cat_cols, verbose=False)

print(grid_search_catboost.best_params_)
print(grid_search_catboost.best_score_)
```

```python
boost_preds_train_simple = grid_search_catboost.predict_proba(features_valid)[:,1]
CBST_roc_auc = roc_auc_score(target_valid, boost_preds_train_simple)

print("CatBoost с подбором гиперпараметров:")
print("Значение AUC-ROC:", CBST_roc_auc)
```
